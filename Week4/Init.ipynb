```python
nparray = np.genfromtxt('./befkbhalderstatkode.csv', delimiter=',', dtype=np.uint, skip_header=1)
print(nparray)
```

```python
neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave', 
       5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst', 
       10: 'Amager Vest', 99: 'Udenfor'}
dd = nparray
for part in neighb:
    mask = (dd[:,0] == 2015) & (dd[:,1] == part)
    print(str(np.sum(dd[mask][:,4])) + ' people in ' + neighb.get(part))
```

```python
import matplotlib.pyplot as plt
yes = {}
for part in neighb:
    mask = (dd[:,0] == 2015) & (dd[:,1] == part)
    yes[neighb.get(part)] = np.sum(dd[mask][:,4])

scity = {k: v for k, v in sorted(yes.items(), key=lambda item: item[1])}

plt.figure(num=None, figsize=(22, 7), dpi=80, facecolor='w', edgecolor='k')
plt.title('People in Copenhagen - 2015', fontsize=20)
plt.bar(scity.keys(), scity.values(), align='center')
plt.ylabel('People', fontsize=14)
plt.xlabel('City part', fontsize=14)
```
